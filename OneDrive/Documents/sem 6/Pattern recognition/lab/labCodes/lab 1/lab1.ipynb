NUMPY

```python
import numpy as np

# 1-D array
arr1 = np.array([1, 2, 3])
print(arr1)
print("\n")

# 2-D array
arr2 = np.array([[1, 2], [3, 4]])
print(arr2)
print("\n")

# 3-D array
arr3 = np.array([[[1, 2], [3, 4]], [[5, 6], [7, 8]]])
print(arr3)
```

```python
print(arr2.ndim)  # Dimensions
print(arr2.size)  # Number of elements
print(arr2.shape)  # Shape of the array
```

```python
arr = np.array([1, 2, 3, 4])
reshaped = arr.reshape(2, 2)
print(reshaped)

flattened = reshaped.flatten()
print(flattened)
```

```python
arr1 = np.array([1, 2, 3])
arr2 = np.array([4, 5, 6])

# Arithmatic Operation
print(arr1 + arr2)
print(arr1 - arr2)
print(arr1 * arr2)
print(arr1 / arr2)
```

```python
arr = np.array([[1, 2, 3], [4, 5, 6]])
print(arr[1, 2]) 
```

```python
# Slicing
print(arr[:, 1])  # Second column

# Masking
mask = arr > 3
print(arr[mask])
```

```python
arr = np.array([1, 2, 3, 4, 5, 6])
reshaped = arr.reshape(2, 3)
print(reshaped)
print("\n")
transposed = reshaped.T
print(transposed)
```

```python
import pandas as pd

# Create a sample DataFrame
data = {
    'Name': ['John', 'Anna', 'Peter', 'Linda', 'Tom'],
    'Age': [28, 24, 35, 32, 40],
    'Country': ['USA', 'UK', 'Australia', 'Germany', 'USA'],
    'Score': [85, 90, 78, 92, 88]
}
df = pd.DataFrame(data)

# Sort the DataFrame by 'Age' in ascending order
df_sorted = df.sort_values(by='Age')

# Filter the DataFrame to include only rows where 'Age' is greater than 30
df_filtered = df[df['Age'] > 30]

# Filter the DataFrame to include only rows where 'Country' is 'USA'
df_filtered_country = df[df['Country'] == 'USA']

# Sort and filter the DataFrame
df_sorted_filtered = df.sort_values(by='Age').loc[df['Age'] > 30]

print("Original DataFrame:")
print(df)
print("\nSorted DataFrame:")
print(df_sorted)
print("\nFiltered DataFrame (Age > 30):")
print(df_filtered)
print("\nFiltered DataFrame (Country == 'USA'):")
print(df_filtered_country)
print("\nSorted and Filtered DataFrame:")
print(df_sorted_filtered)
```

```python
import pandas as pd

df1 = pd.DataFrame({
    'id': [1, 2, 3, 4, 5],
    'name': ['John', 'Mary', 'David', 'Emily', 'Michael']
})

df2 = pd.DataFrame({
    'id': [1, 2, 3, 4, 5],
    'age': [25, 31, 42, 28, 35]
})

# Merge datasets
df_merge = pd.merge(df1, df2, on='id')

# Join datasets
df_inner_join = pd.merge(df1, df2, on='id', how='inner')
df_left_join = pd.merge(df1, df2, on='id', how='left')
df_right_join = pd.merge(df1, df2, on='id', how='right')
df_outer_join = pd.merge(df1, df2, on='id', how='outer')

print("Merged DataFrame:")
print(df_merge)

print("\nInner Join DataFrame:")
print(df_inner_join)

print("\nLeft Join DataFrame:")
print(df_left_join)

print("\nRight Join DataFrame:")
print(df_right_join)

print("\nOuter Join DataFrame:")
print(df_outer_join)
```

```python
import pandas as pd
import numpy as np

# Create a sample dataset
data = {
    'Name': ['John', 'Anna', 'Peter', 'Linda', 'Phil', 'Lucy'],
    'Age': [28, 24, 35, 32, 36, 25],
    'Country': ['USA', 'UK', 'Australia', 'Germany', 'USA', 'UK'],
    'Score': [85, 90, 78, 92, 88, 76]
}
df = pd.DataFrame(data)

# Perform statistical operations
print("Mean of Score:", df['Score'].mean())
print("Median of Score:", df['Score'].median())
print("Mode of Score:", df['Score'].mode().values[0])
print("Standard Deviation of Score:", df['Score'].std())
print("Variance of Score:", df['Score'].var())
print("Correlation between Age and Score:", df['Age'].corr(df['Score']))

# Group by Country and calculate mean Score
print("Mean Score by Country:")
print(df.groupby('Country')['Score'].mean())

# Calculate the 25th, 50th, and 75th percentiles of Score
print("Percentiles of Score:")
print(df['Score'].quantile([0.25, 0.5, 0.75]))
```
